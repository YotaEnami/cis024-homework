```python
# similarity.py

def similarity(inputWord,wordCorpus):
    
    # get list of words that share the relatively same size +/- 1 letter
    
    if len(inputWord) == 0:
        print "Please provide a valid word"
        return "Please enter a word"
    
    lowerWordLen = len(inputWord) - 1
    upperWordLen = len(inputWord) + 1
    
    # get the list of candidate words
    candidateWords = []
    
    for entry in wordCorpus:
        
        if len(entry) >= lowerWordLen and len(entry) <= upperWordLen:
            candidateWords.append(entry)
            
    # perform similarity comparison
    firstTime = True
    bestMatchCount = None
    bestMatchWord = None
    
    print "Input word:",inputWord
    for entry in candidateWords:
        matchCount = 0
        for l1 in entry:
            #print "letter:",l1
            if l1 in inputWord:
                matchCount += 1
                
        if firstTime == True:
            bestMatchCount = matchCount
            bestMatchWord = entry
            firstTime = False
        else:
            if matchCount > bestMatchCount:
                bestMatchCount = matchCount
                bestMatchWord = entry
                
    # display the best match
    return bestMatchWord
```

```python
#index.py
from flask import Flask, render_template, request
import similarity

app = Flask(__name__)

@app.route("/", methods=['GET', 'POST'])

def spellCheck():

    wordlist = open("wordlist.txt")
    wordCorpus = []

    for line in wordlist:
        
        # remove newlines
        line = line.strip().lower()
        
        # get words
        words = line.split(" ")
        
        for word in words:
            if word.isalnum():
                if word not in wordCorpus:
                    wordCorpus.append(word)

    userInput = ""
    result = ""

    if request.method == 'POST':
       	userInput = request.form["userInput"]
        result = similarity.similarity(userInput, wordCorpus)

    return render_template("index.html", userInput = userInput, result = result)

if __name__ == '__main__':
    app.run(host='0.0.0.0',port=80, debug=True)
```
